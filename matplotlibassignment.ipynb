```python
'''Read Total profit of all months and show it using a line plot'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
month_number=filedata['month_number'].tolist()
total_profit=filedata['total_profit'].tolist()
mp.xlabel('Month Number')
mp.ylabel('Total profit')
mp.title('company profit per month')
mp.xticks(month_number)
mp.yticks([100000,200000,300000,400000,500000])
mp.plot(month_number,total_profit)
mp.show()
```

```python
'''Get Total profit of all months and show line plot with the following Style properties'''
'''Generated line plot must include following Style properties: –

Line Style dotted and Line-color should be red
Show legend at the lower right location.
X label name = Month Number
Y label name = Sold units number
Add a circle marker.
Line marker color as read
Line width should be 3'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
month_number=filedata['month_number'].tolist()
total_profit=filedata['total_profit'].tolist()
mp.xlabel('Month Number')
mp.ylabel('Solid units number')
mp.title('company profit per month')
mp.plot(month_number,total_profit,color='red',linewidth='3',marker='o',label='Profit data of last year',linestyle='--')
mp.legend(loc='lower right')
mp.yticks([100000,200000,300000,400000,500000])
mp.xticks(month_number)
mp.show()

```

```python
'''Display the number of units sold per month for each product using multiline plots.
(i.e., Separate Plotline for each product for each product).'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
facecream=filedata['facecream'].tolist()
facewash=filedata['facewash'].tolist()
toothpaste=filedata['toothpaste'].tolist()
bathingsoap=filedata['bathingsoap'].tolist()
shampoo=filedata['shampoo'].tolist()
moisturizer=filedata['moisturizer'].tolist()
month_number=filedata['month_number'].tolist()
mp.xlabel('Month Number')
mp.ylabel('Sales unit in number')
mp.xticks(month_number)
mp.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
mp.title('Sales Data')
mp.plot(month_number,facecream,label='Face Cream Sales Data',color='blue',marker='o',linewidth=3)
mp.plot(month_number,facewash,label='Face Wash Sales Data',color='orange',marker='o',linewidth=3)
mp.plot(month_number,toothpaste,label='Tooth Paste Sales Data',color='green',marker='o',linewidth=3)
mp.plot(month_number,bathingsoap,label='Bathing Soap Sales Data',color='red',marker='o',linewidth=3)
mp.plot(month_number,shampoo,label='Shampoo Sales Data',color='violet',marker='o',linewidth=3)
mp.plot(month_number,moisturizer,label='Moisturizer Sales Data',color='brown',marker='o',linewidth=3)
mp.legend(loc='upper left')
mp.show()
```

```python
''' Read toothpaste sales data of each month and show it using a scatter plot'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
toothpaste=filedata['toothpaste'].tolist()
month_number=filedata['month_number'].tolist()
mp.ylabel('Number of units sold')
mp.xlabel('Month Number')
mp.title('Tooth Paste Sales Data')
mp.xticks(month_number)
mp.grid(b=True, linestyle="--",linewidth= 1)
mp.scatter(month_number,toothpaste,label='Tooth paste Sales data')
mp.legend()
mp.show()
```

```python
'''Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product.
Add a separate bar for each product in the same chart.
'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
facecream=filedata['facecream'].tolist()
facewash=filedata['facewash'].tolist()
month_number=filedata['month_number'].tolist()
mp.yticks([500,1000,1500,2000,2500,3000,3500])
mp.title('Facewash and facecream sales data')
mp.xticks(month_number)
mp.xlabel('Month Number')
mp.ylabel('Sales Unit in Number')
mp.bar([a-0.25 for a in month_number],facecream,label='Face cream Sales data',color='blue',width=0.35)
mp.bar([a+0.25 for a in month_number],facewash,label='Face wash Sales data',color='yellow',width=0.35)
mp.legend(loc='upper left')
mp.grid(True,linewidth=1,linestyle='--')
mp.show()
```

```python
'''Read all product sales data and show it using the stack plot'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
facecream=filedata['facecream'].tolist()
facewash=filedata['facewash'].tolist()
toothpaste=filedata['toothpaste'].tolist()
bathingsoap=filedata['bathingsoap'].tolist()
shampoo=filedata['shampoo'].tolist()
moisturizer=filedata['moisturizer'].tolist()
month_number=filedata['month_number'].tolist()
mp.xlabel('Month Number')
mp.ylabel('Sales unit in number')
mp.title('All product sales data using stack plot')
mp.plot([],[],label='face cream')
mp.plot([],[],label='face wash')
mp.plot([],[],label='toothpaste')
mp.plot([],[],label='bathingsoap')
mp.plot([],[],label='shampoo')
mp.plot([],[],label='moisturizer')

mp.stackplot(month_number,facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer,colors=['pink','blue','red','black','green','yellow'])
mp.legend(loc='upper left')
mp.show()
```

```python
'''Calculate total sale data for last year for each product and show it using a Pie chart'''
import pandas as pd
import matplotlib.pyplot as mp
filedata=pd.read_csv('company_sales_data.csv',delimiter=',')
facecream=filedata['facecream'].tolist()
facewash=filedata['facewash'].tolist()
toothpaste=filedata['toothpaste'].tolist()
bathingsoap=filedata['bathingsoap'].tolist()
shampoo=filedata['shampoo'].tolist()
moisturizer=filedata['moisturizer'].tolist()
month_number=filedata['month_number'].tolist()
sizes=[150,100,200,300,120,100]
label=['Face cream','Face wash','tooth paste','bathing soap','shampoo','moisturizer']
colors1=['blue','orange','green','red','violet','brown']
mp.pie(sizes,colors=colors1,autopct='%1.1f%%',labels=label)
mp.legend(loc='lower right', bbox_to_anchor=(1.25, -0.05))
mp.show()
```

```python
'''Write a NumPy program to check two random arrays are equal or not.'''
import numpy as np
new_array=np.random.random(5)
print(new_array)
new_array2=np.random.random(5)
print(new_array2)
if(np.array_equal(new_array,new_array2)):
    print("Both the arrays are equal")
else:
    print("Both the arrays have different elements")
```

```python
''' Write a NumPy program to create random vector of size 15 and replace the maximum value by -1'''
import numpy as np
new_array1=np.random.random(size=15)
print(new_array1)
max_value=new_array1.max()
print(max_value)
indexmax=np.where(new_array1==max_value)
new_array1[indexmax[0]]=-1
print(new_array1)
```
